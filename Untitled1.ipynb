So it turns out our population is mostly men after all. So let's see if we can figure out if my previous observation is accurate or not.

```python
def barplot_distribution(x,y,hue):
    prop_df = (train[x]
           .groupby(train[hue])
           .value_counts(normalize=True)
           .rename(y)
           .reset_index())

    sns.barplot(x=x, y=y, hue=hue, data=prop_df).set_title('Distribution of {} among {}s'.format(hue,x))
```

```python
barplot_distribution('Loan_Status','%','Gender')
```

##### So it seems that Gender has no influence in the loan decisions after all.

```python
barplot_distribution('Education','%','Loan_Status')
```
