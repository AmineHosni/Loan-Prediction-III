Most of our female observations are not married while most of our male ones are Married
So we're going to use this to try to guess the Gender based on their Married Status

```python
def impute_Gender(cols):
    Gender = cols[0]
    Married = cols[1]
    if pd.isnull(Gender):
        if Married == 'Yes':
            return 'Male'
        else:
            return 'Female'
    else:
        return Gender
```

```python
train['Gender'] = train[['Gender','Married']].apply(impute_Gender,axis=1)
```

#### We noticed almost everyone with no Credit_History has no Loan Status 
#### So we're gonna go ahead and use this feature due to its high correlation with our Credit History

```python
def impute_Credit_History_no_Loans(cols):
    Credit_History = cols[0]
    Loan_Status = cols[1]
    if pd.isnull(Credit_History):
        if Loan_Status == 'N':
            return 0.0
        else:
            return 1.0
    else:
        return Credit_History
```

```python
train['Credit_History'] = train[['Credit_History','Loan_Status']].apply(impute_Credit_History_no_Loans,axis=1)
```

```python
def coding(col, codeDict):
  colCoded = pd.Series(col, copy=True)
  for key, value in codeDict.items():
    colCoded.replace(key, value, inplace=True)
  return colCoded
```

```python
print('Before Coding:')
print(train_copy['Loan_Status'].value_counts())
train_copy["Loan_Status"] = coding(train_copy["Loan_Status"], {'N':0,'Y':1})
print('After Coding:')
print(train_copy['Loan_Status'].value_counts())
```

```python
print('Before Coding:')
print(train_copy['Married'].value_counts())
train_copy["Married"] = coding(train_copy["Married"], {'No':0,'Yes':1})
print('After Coding:')
print(train_copy['Married'].value_counts())
```

```python
print('Before Coding:')
print(train_copy['Self_Employed'].value_counts())
train_copy["Self_Employed"] = coding(train_copy["Self_Employed"], {'No':0,'Yes':1})
print('After Coding:')
print(train_copy['Self_Employed'].value_counts())
```

```python
print('Before Coding:')
print(train_copy['Gender'].value_counts())
train_copy["Gender"] = coding(train_copy["Gender"], {'Female':0,'Male':1})
print('After Coding:')
print(train_copy['Gender'].value_counts())
```

```python
print('Before Coding:')
print(train_copy['Education'].value_counts())
train_copy["Education"] = coding(train_copy["Education"], {'Not Graduate':0,'Graduate':1})
print('After Coding:')
print(train_copy['Education'].value_counts())
```
